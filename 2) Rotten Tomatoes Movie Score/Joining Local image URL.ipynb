```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import wordcloud
from tqdm import tqdm_notebook as progressbar
import os
%matplotlib inline
```

```python
df = pd.read_csv('full_movies_reviews.csv')
df.head(3)
```

```python
# df['local_image'] = 
image_folder = 'bestofrt_posters'
local_url_list = os.listdir(image_folder)
reference_list = []
for url in local_url_list:
    ranking = url.split('_')[0]
    d = {'ranking': int(ranking), 'url':url}
    reference_list.append(d)
```

```python
df_local_url = pd.DataFrame(reference_list, columns=['ranking','url'])
```

```python
df_local_url.url = image_folder+'/'+df_local_url.url
```

```python
df = pd.merge(df,df_local_url, on='ranking')
```

```python
df.rename(columns={'url':'local_url'})
```

```python
df.to_csv('final_dataframe.csv', index=False)
```
